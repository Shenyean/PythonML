# Preping the Dataframe

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
pricepromo = df[['Price','Promotion']]
sales = df['Sales'] # This is the same as
sales = df.Sales

```

# training the model
We will test the linear model using a loop of different randsom state. the model with the Rsquare which meets the condition set by me > 0.9200 will be save and used to predict future data. 

```python
for i in range (0,101):
    x_train, x_test, y_train, y_test = train_test_split(pricepromo, sales, test_size=0.2, random_state = i)
    from sklearn.linear_model import LinearRegression
    linReg = LinearRegression()
    linReg.fit(x_train, y_train)
    y_hat = linReg.predict(x_test)
    from sklearn.metrics import mean_squared_error, r2_score
    r2_score(y_test, y_hat)
    print(r2_score(y_test, y_hat))
    if r2_score(y_test, y_hat)>0.9200:
        from joblib import dump, load
        dump(linReg, 'linreg.joblib') 
        break
```

```python
linReg2 = load('LinReg.joblib') 
y_hat = linReg2.predict(x_test)
print(y_hat)
#print(x_test)
print(y_test)
r2_score(y_test, y_hat)
```
